```python
import pandas as pd

df = pd.read_csv("student-mat.csv",sep=";")
print(df.head())  # .head() will give the information of first five cells.
```

```python
print(df.isnull().sum())   # isnull() tells that where the cells have null values and sum() gives the exact no of null values.
```

```python
print(df.dtypes)  #.dtypes() gives the data types of all the columns.
```

```python
print(df.shape) # .shape gives the total no of rows and column.
```

```python
print(df.drop_duplicates())  #drop_duplicates() removes all the duplicate rows from the data frame
```

```python
rowAvg = df["G3"].mean()    # .mean()  this will give the mean of the any column.
print(rowAvg)
```

```python
print((df["G3"]>15).sum())  # this will give the total no of students who scored more than 15 in G3.
```

```python
Cor = df["studytime"].corr(df["G3"])
print(Cor)  # the output is 0.0978 which is very close to 0 that means there is almost no relationship between the two column.
```

```python
avgGrade = df.groupby("sex")["G3"].mean()
print(avgGrade.idxmax())    # The avg marks of both gender will be stored in avgGrade than the idxmax() will give the gender of maximum average grade among them.
```

```python
import matplotlib.pyplot as plt               
plt.hist(df["G3"],bins = 10,edgecolor = "Black")        #plt.hist will create histogram , bins will divide the bar like here bins is 10 so 10 bar ,and title adds title for the histogram
plt.title("Final Grades of G3")
plt.show()
```

```python
avg_marks = df.groupby("studytime")["G3"].mean()

plt.bar(avg_marks.index, avg_marks.values)
plt.title("Average Final Grade by Study Time")
plt.show()
```
